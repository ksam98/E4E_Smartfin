```python
#Purisa Jasmine Simmons
#July 18, 2018
#Calculating accurate vertical acceleration from accelerometer readings and gyroscope readings.

#Based on: “A Comparison of Methods for Determining Significant Wave Heights”
#(https://journals.ametsoc.org/doi/pdf/10.1175/2010JTECHO724.1)

#Overview:
#1.Calculate vertical accelerations using Method V (includes gyroscope)
#2.Filter to remove outliers in acceleration data
#3.FFT of filtered, 4-Hz acceleration data using Kaiser-Bessel window
#4.Frequency-domain filter to remove low-frequency noise
#5.Noise-corrected acceleration spectra is converted into displacement spectra by dividing by the frequency to the fourth power, heave RAO used by NDBC applied
#6.Significant wave height, peak period, and mean wave were determined from the displacement spectra using definitions provided on the NDBC website: http:// www.ndbc.noaa.gov/

%matplotlib inline

import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

from plotly import tools
import plotly.offline
import plotly.graph_objs as go

import math
import re
import csv

#f = open("Motion_CE_14814.CSV")
#csv_f = csv.reader(f)
#for row in csv_f:
#    print row
    
print('Running Acceleration Calculation Algorithm:')

#Reading data from filename_r
filename_r = "Motion_CE_14814.CSV"
read_file = open(filename_r, "r")


#Initalize lists
imua1_list = []
imua2_list = []
imua3_list = []

imug1_list = []
imug2_list = []
imug3_list = []

time_o_list = []
time_e_list = []

pitch_list = []
roll_list = []

vert_a_list = []


#Calculate time_offsets
with open(filename_r, 'r') as f: 
    for line in f:
        #print line     #helps with debugging
        str_array = line.split(',')  #separates each line into an array on commas

        #-------Calculating Time Offset--------#
        if str_array[1] == "Time":
            t1 = 0
            t2 = 0
            time_o_list.append(0)  #initialize time_list with 0
            time_e_list.append(0)

        else:
            t2 = str_array[1]    #Using "Time" Column instead of "UTC" column

            if (t2 != 0 and t1 != 0 and str_array[2] != "N/A"):
                t_out = float(t2) - float(t1) #measured in secs
                t_out = (t_out*(10**-3))

                last_ti = len(time_o_list) - 1 #index of last time_offset
                last_te = time_e_list[last_ti] #last time_elapsed
                time_e_list.append(last_te + t_out)  #last time_elapsed + time_offset
                time_o_list.append(t_out)

        
            if (str_array[2] != "N/A" and str_array[3] != "N/A" and \
                str_array[4] != "N/A" and str_array[2] != "IMU A1"):


                #Scale raw to get correct units in m/s^2
                g_const = 512   #constant for gravity: 512 

                #a1 = -float(str_array[2])/g_const  #x-axis (horizontal direction 1, forwards/backwards)
                a1 = float(str_array[2])/g_const  #x-axis (horizontal direction 1, forwards/backwards)
                a2 = float(str_array[3])/g_const  #y-axis, affected by gravity (vertical)
                a3 = float(str_array[4])/g_const  #z-axis (horizontal direction 2)
        
                d_const = 8.2 #constant for raw units to degrees
        
                g1 = float(str_array[5])/d_const  #roll
                g2 = float(str_array[6])/d_const  #yaw
                g3 = float(str_array[7])/d_const  #pitch

                #print(("ax: %f") % ax) 
                #print(("ay: %f") % ay)
                #print(("az: %f") % az)
        
                
                imua1_list.append(a1)
                imua2_list.append(a2)
                imua3_list.append(a3)
        
                imug1_list.append(g1)
                imug2_list.append(g2)
                imug3_list.append(g3)
                

        #Reset t1 and t2 after t_out is calculated
        t1 = t2
        t2 = 0

#------Here after lists have been created---------
#Check if all lists are the same length, they should be:
if (len(imua1_list) != len(imug1_list) and len(imua1_list) != len(time_o_list)):
    print("Error! List lengths do not match.")
else:
    
    g = -9.80665 #1g = 9.80665 m/s^2
    #g = 1
    
    for a1,a2,a3,g1,g2,g3 in zip(imua1_list, imua2_list, imua3_list, imug1_list, imug2_list, imug3_list): 
        
        theta = g3 #pitch
        gamma = -g1 #roll
        
        #i = -math.sin(theta) #corresponds to a3 in the paper
        #j = -math.sin(gamma)*math.cos(theta) #b3 in the paper
        #k = -math.cos(gamma)*math.cos(theta) #c3 in the paper
        
        i = -math.sin(theta)
        j = math.sin(gamma)*math.cos(theta)
        k = math.cos(gamma)*math.cos(theta)
        
        a1 = a1*g
        a2 = a2*g
        a3 = a3*g
        
        Az = g*(1-(i*a1 + j*a3 + k*a2))
        vert_a_list.append(Az)
        
        #print a1
        #print a2
        #print a3 
        
        #print(Az)
        
    time = np.array(time_e_list)
    acc = np.array(vert_a_list)
        
        
    plt.plot(time, acc)
    plt.title(filename_r)
    plt.xlabel('Time [s]')
    plt.ylabel('Vertical acceleration [m/s^2]')
    plt.show
   
```
